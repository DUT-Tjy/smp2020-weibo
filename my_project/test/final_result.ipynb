```python
import pandas as pd
import json
```

```python
test_df = pd.DataFrame(columns=['id','label_0','label_1','label_2','label_3','label_4','label_5'])
test_df = pd.read_csv('fine_base/sub_0.csv') 
test_origin = pd.read_csv('fine_base/sub_0.csv') 
name_list = ['fine_base','fine_gru','fine_last2embedding']
for i in range(len(name_list)):
    test_df_0 = pd.read_csv(name_list[i]+'/sub_0.csv') 
    test_df_1 = pd.read_csv(name_list[i]+'/sub_1.csv') 
    test_df_2 = pd.read_csv(name_list[i]+'/sub_2.csv') 
    test_df_3 = pd.read_csv(name_list[i]+'/sub_3.csv') 
    test_df_4 = pd.read_csv(name_list[i]+'/sub_4.csv')
    test_df = test_df + test_df_0 + test_df_2 + test_df_2 + test_df_3 + test_df_4
test_df = test_df - test_origin
```

```python
test_df
```

```python
name = 'label_'
for i in range(len(test_df)):
    max_index = 0
    max_num = test_df['label_0'][i]
    for j in range(6):
        name_new = name + str(j)
        if test_df[name_new][i] > max_num:
            max_index = j
            max_num = test_df[name_new][i]
    for j in range(6):
        name_new = name + str(j)
        test_df[name_new][i] = 0
    name_new = name + str(max_index)
    test_df[name_new][i] = 1
```

```python
test_df
```

```python
result_0 = test_df[:50000]
result_1 = test_df[50000:]
```

```python
result_0['result'] = 'happy'
name = 'label_'
for i in range(len(result_0)):
    final_index = 0
    for j in range(6):
        name_new = name + str(j)
        if result_0[name_new][i] == 1:
            final_index = j
    if final_index == 0:
        result_0['result'][i] = 'happy'
    elif final_index == 1:
        result_0['result'][i] = 'angry'
    elif final_index == 2:
        result_0['result'][i] = 'neural'
    elif final_index == 3:
        result_0['result'][i] = 'sad'
    elif final_index == 4:
        result_0['result'][i] = 'surprise'
    elif final_index == 5:
        result_0['result'][i] = 'fear'
```

```python
result_1['result'] = 'happy'
name = 'label_'
for i in range(len(result_1)):
    final_index = 0
    for j in range(6):
        name_new = name + str(j)
        if result_1[name_new][50000+i] == 1:
            final_index = j
    if final_index == 0:
        result_1['result'][i] = 'happy'
    elif final_index == 1:
        result_1['result'][i] = 'angry'
    elif final_index == 2:
        result_1['result'][i] = 'neural'
    elif final_index == 3:
        result_1['result'][i] = 'sad'
    elif final_index == 4:
        result_1['result'][i] = 'surprise'
    elif final_index == 5:
        result_1['result'][i] = 'fear'
```

```python
data_all = []
for i in range(len(result_0)):
    data_final = {}
    data_final['id'] = i + 1
    data_final['label'] = result_0['result'][i]
    data_all.append(data_final)
json.dump(data_all, open("usual_result.txt", "w", encoding="utf8"))
```

```python
data_all = []
for i in range(len(result_1)):
    data_final = {}
    data_final['id'] = i + 1
    data_final['label'] = result_1['result'][i]
    data_all.append(data_final)
json.dump(data_all, open("virus_result.txt", "w", encoding="utf8"))
```
